```python
import csv
import requests
from bs4 import BeautifulSoup

def scrape_amazon_products(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the product containers
    products = soup.find_all('div', {'data-component-type': 's-search-result'})

    # Create a list to store the product details
    product_details = []

    for product in products:
        # Extract the product name
        name = product.find('span', {'class': 'a-size-base-plus'}).text.strip()

        # Extract the product price
        price_element = product.find('span', {'class': 'a-offscreen'})
        price = price_element.text.strip() if price_element else 'N/A'

        # Extract the product rating
        rating_element = product.find('span', {'class': 'a-icon-alt'})
        rating = rating_element.text.strip() if rating_element else 'N/A'

        # Extract the seller name
        seller_element = product.find('span', {'class': 'a-size-small'})
        seller = seller_element.text.strip() if seller_element else 'Out of stock'

        # Append the product details to the list
        product_details.append([name, price, rating, seller])

    return product_details

def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Product Name', 'Price', 'Rating', 'Seller Name'])

        for row in data:
            writer.writerow(row)

# URL of the Amazon page
url = 'https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar'

# Scrape the product details
product_details = scrape_amazon_products(url)

# Save the product details to a CSV file
save_to_csv(product_details, 'amazon_products.csv')
```

```python
#this code works

import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

def scrape_amazon_products(url):
    options = Options()
    options.headless = True

    # Set path to your ChromeDriver executable
    chrome_driver_path = '/path/to/chromedriver'

    # Create a new ChromeDriver service
    service = Service(chrome_driver_path)

    # Launch the WebDriver
    driver = webdriver.Chrome(service=service, options=options)

    # Load the Amazon page
    driver.get(url)

    # Extract the HTML content
    html = driver.page_source

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Find all the product containers
    products = soup.find_all('div', {'data-component-type': 's-search-result'})

    # Create a list to store the product details
    product_details = []

    for product in products:
        # Extract the product name
        name_element = product.find('span', {'class': 'a-size-base-plus'})
        name = name_element.text.strip() if name_element else 'N/A'

        # Extract the product price
        price_element = product.find('span', {'class': 'a-offscreen'})
        price = price_element.text.strip() if price_element else 'N/A'

        # Extract the product rating
        rating_element = product.find('span', {'class': 'a-icon-alt'})
        rating = rating_element.text.strip() if rating_element else 'N/A'

        # Get the product URL
        product_url = product.find('a', {'class': 'a-link-normal'})['href']
        product_url = 'https://www.amazon.in' + product_url

        # Open the product page
        driver.get(product_url)

        # Extract the seller name from the product page
        seller_element = driver.find_element(By.ID, 'bylineInfo')
        seller = seller_element.text.strip() if seller_element else 'Out of stock'

        # Append the product details to the list
        product_details.append([name, price, rating, seller])

    # Close the WebDriver
    driver.quit()

    return product_details

def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Product Name', 'Price', 'Rating', 'Seller Name'])

        for row in data:
            writer.writerow(row)

# URL of the Amazon page
url = 'https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar'

# Scrape the product details
product_details = scrape_amazon_products(url)

# Save the product details to a CSV file
save_to_csv(product_details, 'amazon_products.csv')
```
